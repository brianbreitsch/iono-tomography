# Introduction to Tomography

Tomography is a form of image reconstruction which uses projections through the image as its source of information.

Quoth Wikipedia:

<div style="width:80%; margin: 0 auto;">
*"Tomography refers to imaging by sections or sectioning, through the use of any kind of penetrating wave."*
</div>

For a foundation on the ideas and history behind modern tomographic methods, the [Wikipedia page on Tomography] is a good place to start.


[Wikipedia page on Tomography]:http://en.wikipedia.org/wiki/Tomography

### objectives

In this notebook, we will introduce the concept of tomography, as well as some helpful libraries and programming tools that we will use in successive notebooks.

The ***Shepp-Logan phantom*** is the de-facto test image for many in the field of tomography. This image mimics density variations of a brain tumor. Its originators, Larry Shepp and Benjamin F. Logan, devised it for use in their 1974 paper *The Fourier Reconstruction of a Head Section*, which studied tomography for medical applications.

Since it's so iconic, we'll use it for this introduction. However, we should note that it doesn't really resemble the ionosphereic images we'll be reconstructing later on. As a coding exercise, let's load it and display it using **matplotlib**!

```python
import scipy.ndimage as ndimage
import matplotlib.pyplot as plt
%matplotlib inline

img = ndimage.imread('../img/shepp_logan_phantom.png', flatten=True)
plt.imshow(img, cmap='gray', interpolation='nearest')
plt.show()
```

### software

There are several software packages available for tomographic imaging:

Another matlab implementation of tomographic reconstruction using filtered back-projection. Medical-oriented.

*  ASTRA Tomography Toolbox

The "All Scale Tomographic Reconstruction Antwerp", or ASTRA, is a tomography c++ library and MATLAB toolbox developed by VisionLab at the University of Antwerp. Cited as a "high-performance GPU MATLAB toolbox for 2D and 3D tomography", this toolbox is possibly the most feature-rich tomography toolkit available. It is open-source and hosted at the [ASTRA sourceforge page]. In addition to the MATLAB toolbox, there is a separate project, *pyastratoolbox*, that offers a Python interface to the library. It seems to be targeted more towards dense projection geommetries and smaller scales than we would like to accomplish with ionosphere tomography.

[ASTRA sourceforge page]:http://sourceforge.net/p/astra-toolbox/wiki/Home/

- [Peter Toft's Thesis]

This thesis contains example reconstruction techniques and some references to implementation. The focus is medical--although some techniques are applicable in the ionosphere tomography domain.

## working with 2d images

```python
import skimage.transform as skit
```

The `scikit-image` module has some useful functions for transforming and manipulating 2d images. It contains functions for performing affine and projection transforms, as well as for computing forward and inverse *Radon transform*. Here's an example of using its Radon transform capabilities.

```python
sinogram = skit.radon(img)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(sinogram)
plt.show()
```

```python
reconstructed_img = skit.iradon_sart(sinogram)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(reconstructed_img)
plt.show()
```

The **Radon transform** is named after the German mathematician Johann Radon (1885-1917) who discovered its inverse transformation. One form of tomography manifests itself through the inverse Radon transform, where a complete set of projections through an imaging region are used to reconstruct an image.
