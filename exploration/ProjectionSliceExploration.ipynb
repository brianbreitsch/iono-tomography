### The Hankel Tranform

forward:

$$
S(f) = 2\pi\int_{0}^{\infty} ts(t)J_0(2\pi f t) \ dt
$$

inverse:

$$
s(t) = 2\pi\int_{0}^{\infty} tS(t)J_0(2\pi f t) \ df
$$

Transform of 1D function for 2D radially symmetric function to 1D function of 2D radially symmetric Fourier transform.

We iterate through the different image slices determined by $\theta$. For each slice, we use the ndimage function map_coordinates to create profiles of the image. We also use projection-slice theorem (which states that $P_\theta(f) = S(f\cos\theta,f\sin\theta)$--that is that the Fourier transform of a projection slice is simply a slice of the 2D Fourier transform of our image function $s$) in order to produce projections of the image at each $\theta$.

```python
imag = np.imag(img_freq)
real = np.real(img_freq)

N = 31
res = 500
plot_proj_lines = True

if plot_proj_lines:
    fig = plt.figure()
    img_ax = fig.add_subplot(1,2,1)
    img_ax.imshow(img)
    ax.set_xticks([]); ax.set_yticks([]); ax.set_aspect(1.)
    ax = fig.add_subplot(1,2,2)
    img_freq = np.fft.fftshift(np.fft.fft2(img))
    ax.imshow(np.abs(img_freq))
    ax.set_xticks([]); ax.set_yticks([]); ax.set_aspect(1.)

fig = plt.figure()

for theta, i in zip(np.linspace(0.,np.pi,N), range(N)):
    x = np.linspace(x0-r*np.cos(theta),x0+r*np.cos(theta),res).reshape((res,1))
    y = np.linspace(y0-r*np.sin(theta),y0+r*np.sin(theta),res).reshape((res,1))
    
    coords = np.concatenate((x,y),axis=1).T
    
    line = ndimage.map_coordinates(img, coords, order=1)
    proj_real = ndimage.map_coordinates(real, coords, order=1)
    proj_imag = ndimage.map_coordinates(imag, coords, order=1)
    proj = proj_real + 1j*proj_imag
    proj = np.fft.ifft(np.fft.fftshift(proj))
    proj = np.abs(proj)
    
    if plot_proj_lines:
        img_ax.plot(x,y,color=(i*.5/N,.7,i*1./N))
    
    ax = fig.add_subplot(N,2,1+2*i)
    ax.plot(line)
    ax.set_xticks([]); ax.set_yticks([])
    
    ax = fig.add_subplot(N,2,2+2*i)
    ax.plot(proj)
    ax.set_xticks([]); ax.set_yticks([])
    
plt.show()
```
