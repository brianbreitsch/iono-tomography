# Radon Exploration

The purpose of this notebook is to explore the Radon transform in a discrete context, specifically the implementation used in *scikit-image*'s *transform* module. We will:

- generate images that provide and intuition behind finite back projection (FBP) reconstruction
- explore the effect of upsampling and downsampling projection data before and after applying forward/inverse Radon transform by various methods
- meditate on the implications for development of a fully general algorithm

```python
import numpy as np
from numpy import pi
import scipy as sp
import scipy.ndimage as ndimage
import scipy.sparse as sparse
import scipy.signal as signal
import matplotlib.pyplot as plt
import skimage #.transform import radon, iradon, _warps, _warps_cy
%matplotlib inline
```

```python
img = ndimage.imread('../img/shepp_logan_phantom.png', flatten=True)
plt.subplot(131)
plt.imshow(img, interpolation='nearest')
plt.subplot(132)
plt.imshow(sp.misc.imresize(sp.misc.ascent(),(32,32)))
plt.subplot(133)
plt.imshow(sp.misc.imresize(sp.misc.face(True),(32,32)), interpolation='nearest')
plt.show()
```

```python
def circle_mask(img, radius=min(img.shape)//2):
    c0, c1 = np.ogrid[0:img.shape[0], 0:img.shape[1]]
    reconstruction_circle = ((c0 - img.shape[0] // 2)**2
                                 + (c1 - img.shape[1] // 2)**2) <= radius**2
    return img * reconstruction_circle
```

```python
def projmtx2(theta, N_proj, img_shape):
    N_theta = len(theta)
    N_rows, N_cols = img_shape
    N_img = N_rows*N_cols
    A = sparse.lil_matrix((N_theta*N_proj,N_img))
    
    delta = np.zeros(img_shape)
    proj = np.zeros((N_theta,N_proj))
    
    for n in range(N_img):
        # move delta function across columns first
        delta[n//N_rows,n%N_cols] = 1.
        for i, t in enumerate(theta):
            proj[i,:] = sp.signal.resample(np.sum(ndimage.rotate(delta,t,reshape=False),axis=0),N_proj)
        A[:,n] = proj.reshape((N_theta*N_proj,1), order='c')
        delta[n//N_rows,n%N_cols] = 0.
    return A
    
```

```python
def radon2(img, theta, N_proj):
    N_theta = len(theta)
    N_rows, N_cols = img.shape
    N_img = N_rows*N_cols
    proj = np.zeros((N_theta,N_proj))
    
    for i, t in enumerate(theta):
        proj[i,:] = sp.signal.resample(np.sum(ndimage.rotate(img,t,reshape=False),axis=0),N_proj)
    
    return proj
```

```python
%%time
N_theta = N_proj = N_rows = N_cols = 19
img = ndimage.imread('../img/I.png', flatten=True)
img = 1. - img
img = sp.misc.imresize(img, (N_rows,N_cols))
img = circle_mask(img, radius=N_cols//2.05)
theta = np.linspace(0.,180.,N_theta)
A = projmtx2(theta, N_proj, img.shape)

proj = radon2(img, theta, N_proj)
proj2 = A.dot(img.reshape((N_rows*N_cols,1))).reshape((N_theta,N_proj))
adjoint_bp = A.T.dot(proj.reshape((N_proj*N_theta,1), order='c')).reshape((N_rows,N_cols))
adjoint_bp2 = A.T.dot(proj2.reshape((N_proj*N_theta,1), order='c')).reshape((N_rows,N_cols))

plt.subplot(241)
plt.imshow(img, interpolation='nearest')
plt.xticks([]); plt.yticks([])

plt.subplot(242)
plt.imshow(proj, interpolation='nearest')
plt.xticks([]); plt.yticks([])

plt.subplot(243)
plt.imshow(proj2, interpolation='nearest')
plt.xticks([]); plt.yticks([])

plt.subplot(244)
plt.imshow(A.todense(), interpolation='nearest')
plt.xticks([]); plt.yticks([])

plt.subplot(246)
plt.imshow(adjoint_bp, interpolation='nearest')
plt.xticks([]); plt.yticks([])

plt.subplot(247)
plt.imshow(adjoint_bp2, interpolation='nearest')
plt.xticks([]); plt.yticks([])

# .844, 2.5, 29.4
plt.show()
```

Above we see the results of our new Radon transform and projection matrix methods *radon2* and *projmtx2*. The middle images represent the projections (top) and adjoint reconstructions (bottom). We create the projection matrix on the left side using the *radon2* method, and its corresponding adjoint reconstruction. The projection matrix on the right side comes from transforming the original image using our projection matrix. We see that the results for both projection matrices and reconstructions are almost identical. The right side projection matrix may be slightly sharper.

### Question:
#### What happens when we repeatedly apply our transformation matrix to an image?

The concept is as follows:

1) take and image and generate an appropriate tranformation matrix
2) apply the transformation matrix to the image
3) apply adjoint reconstruction to the projection
4) repeat steps 2 and 3 indefinitely

Mathematically, we write:

$$s^N = A^TAA^TAA^T...A^TAs $$

where the term $A^TA$ appears $N$ times.

Let's see!

```python
N_theta = N_proj = N_rows = N_cols = 32
img = ndimage.imread('../img/H.png', flatten=True)
img = 1. - img
img = sp.misc.imresize(img, (N_rows,N_cols))
img = circle_mask(img, radius=N_cols//2.05)
theta = np.linspace(0.,180.,N_theta)
A = projmtx2(theta, N_proj, img.shape)

new_img = img.reshape((N_rows*N_cols,1))

N = 14
for i in range(1,N+1):
    proj = A.dot(new_img)
    new_img = A.T.dot(proj)
    plt.subplot(1,N,i)
    plt.imshow(new_img.reshape((N_rows,N_cols)), interpolation='nearest')
    plt.xticks([]); plt.yticks([])
```

It blurs and quickly approaches a steady state as energy is added to the image.

Now lets observe how adjoint reconstruction behaves when only using a partial number of angles.

```python
%pdb
N_theta = N_proj = N_rows = N_cols = 32
img = ndimage.imread('../img/H.png', flatten=True)
img = 1. - img
img = sp.misc.imresize(img, (N_rows,N_cols))
img = circle_mask(img, radius=N_cols//2.05)
theta = np.linspace(0.,180.,N_theta)
A = projmtx2(theta, N_proj, img.shape)

proj = A.dot(img.reshape((N_rows*N_cols,1))).reshape((N_theta,N_proj))

N_angles = np.arange(1,N_theta,5)
for i, N in enumerate(N_angles):
    adjoint_bp = A[0:N*N_proj,:].T.dot(proj[0:N,:].reshape((N_proj*N,1), order='c')).reshape((N_rows,N_cols))
    plt.subplot(1,len(N_angles),i+1)
    plt.imshow(adjoint_bp, interpolation='nearest')
    plt.title('{0} angles'.format(N))
    plt.xticks([]); plt.yticks([])

plt.show()
```

The progression gives us a glimpse of the intuition behind how the reconstruction process works. We see the image become more clear we use more and more projection slices.

Now we want to test non-uniform dimensions in using our *projmtx2* and *radon2* functions. This was a problem in our last implementation, whose root cause was the use of the *radon* function from scikit-image.

```python
%pdb
N_rows = N_cols = 32
N_theta = N_proj = 32
img = ndimage.imread('../img/H.png', flatten=True); img = 1. - img
img = sp.misc.imresize(img, (N_rows,N_cols))
img = circle_mask(img, radius=N_cols//2.05)
theta = np.linspace(0.,180.,N_theta)
A = projmtx2(theta, N_proj, img.shape)

proj = A.dot(img.reshape((N_rows*N_cols,1)))

skip = [1, 2, 3, 4, 5, 6, 7, 8]
for i, N in enumerate(skip):
    adjoint_bp = A[::N,:].T.dot(proj[::N,:]).reshape((N_rows,N_cols))
    plt.subplot(1,len(skip),i+1)
    plt.imshow(adjoint_bp, interpolation='nearest')
    plt.title('skip {0}'.format(N))
    plt.xticks([]); plt.yticks([])

plt.show()
```

```python
%pdb
N_rows = N_cols = 32
img = ndimage.imread('../img/H.png', flatten=True); img = 1. - img
img = sp.misc.imresize(img, (N_rows,N_cols))
img = circle_mask(img, radius=N_cols//2.05)

N_theta_list = [16, 21, 24, 32]
N_proj_list = [8, 12, 16, 24]
for i, (N_theta, N_proj) in enumerate(zip(N_theta_list,N_proj_list)):
    theta = np.linspace(0.,180.,N_theta)
    A = projmtx2(theta, N_proj, img.shape)
    proj = A.dot(img.reshape((N_rows*N_cols,1)))
    adjoint_bp = A.T.dot(proj).reshape((N_rows,N_cols))
    plt.subplot(1,len(N_proj_list),i+1)
    plt.imshow(adjoint_bp, interpolation='nearest')
    plt.title('{0} angles, {1} projs'.format(N_theta,N_proj))
    plt.xticks([]); plt.yticks([])

plt.show()
```

Awesome--non-uniform dimensions for number of projections and projection angles is working! Clearly, the more projection angles and projections the better the reconstruction. Now let's explore the relative importance/difference between number of projections versus projection angles.

```python
%pdb
N_rows = N_cols = 32
img = ndimage.imread('../img/shepp_logan_phantom.png', flatten=True)
img = sp.misc.imresize(img, (N_rows,N_cols))
img = circle_mask(img, radius=N_cols//2.05)

N_theta_list = [3, 7, 11, 16, 24, 32]
N_proj_list = [3, 7, 11, 16, 24, 32]
for i, N_theta in enumerate(N_theta_list):
    for j, N_proj in enumerate(N_proj_list):
        theta = np.linspace(0.,180.,N_theta)
        A = projmtx2(theta, N_proj, img.shape)
        proj = A.dot(img.reshape((N_rows*N_cols,1)))
        adjoint_bp = A.T.dot(proj).reshape((N_rows,N_cols))
        plt.subplot(len(N_theta_list),len(N_proj_list),j+i*len(N_proj_list)+1)
        plt.imshow(adjoint_bp, interpolation='nearest')
        plt.title('{0}, {1}'.format(N_theta,N_proj))
        plt.xticks([]); plt.yticks([])

plt.show()
```

We see that adjoint reconstruction becomes sharper with more projections and becomes less noisy/clearer with more angles.

Now for a quick test of how our algorithm handles non-uniform image dimensions.

```python
%%time
N_rows, N_cols = 32, 24
img = ndimage.imread('../img/shepp_logan_phantom.png', flatten=True)
img = sp.misc.imresize(img, (N_rows,N_cols))
img = circle_mask(img, radius=N_cols//2.05)

N_theta = N_proj = 19
theta = np.linspace(0.,180.,N_theta)
A = projmtx2(theta, N_proj, img.shape)
proj = A.dot(img.reshape((N_rows*N_cols,1)))
adjoint_bp = A.T.dot(proj).reshape((N_rows,N_cols))

plt.subplot(1,3,1)
plt.imshow(img, interpolation='nearest')
plt.xticks([]); plt.yticks([])

plt.subplot(1,3,2)
plt.imshow(adjoint_bp, interpolation='nearest')
plt.xticks([]); plt.yticks([])

plt.show()
```
